Here the web scraping process comes to an end <br>
But we do need to save all this scraped data somewhere right?

That's why we imported that `import csv` package.
- We will be adding header 
- We will be adding the data followed by the header

```python
# defining the header variable with header names
header = ['name','price','rating','review']
```

```python
# writing the file
with open('ev_scraped_data.csv', 'w', encoding= 'UTF-8', newline = '') as file:
    writer = csv.writer(file)
    # writing the header
    writer.writerow(header)
    # writing page1 data
    for name, price, rating, review in zip(page1_name, page1_price, page1_rating, page1_review):
        writer.writerow([name.text,price.text,rating.text, review.text])
    # writing page2 data
    for name, price, rating, review in zip(page2_name, page2_price, page2_rating, page2_review):
        writer.writerow([name.text,price.text,rating.text, review.text])
    # writing page3 data
    for name, price, rating, review in zip(page3_name, page3_price, page3_rating, page3_review):
        writer.writerow([name.text,price.text,rating.text, review.text])
    # writing page4 data
    for name, price, rating, review in zip(page4_name, page4_price, page4_rating, page4_review):
        writer.writerow([name.text,price.text,rating.text, review.text])
    # writing page5 data
    for name, price, rating, review in zip(page5_name, page5_price, page5_rating, page5_review):
        writer.writerow([name.text,price.text,rating.text, review.text])
#closing the file
file.close()
```
